# 3. Features engineering

```python
# Removing useless info columns

dfs =  [df_products, df_orders, df_reviews, df_items]

for df in dfs:

  df.drop(columns = ['Unnamed: 0'], inplace = True)
  
df_products.dropna(inplace = True)
```

```python
# Creating new columns

df_items['tot_price'] = df_items['order_item_id'] * df_items['price']
df_products['density'] = df_products['product_weight_g'] / (df_products['product_length_cm'] * df_products['product_height_cm'] * df_products['product_width_cm'])
```

```python
# Merging dataframes

df1 = df_items.merge(df_reviews, on = 'order_id')
df2 = df1.merge(df_orders, on = 'order_id')
df3 = df2.merge(df_products, on = 'product_id')
```

```python
df3.head()
```

```python
# Wheiging reviews by relevance (2 if the product was delivered, 1 if it wasn't). We've decided that the reviews done by people who hadn't received the product yet are not 
# trustworthy, thus decreasing it's value when compared to the ones who actually received the product

df3['order_status'] = df3['order_status'].apply(lambda x: 2 if x == 'delivered' else 1)

df3['review_score'] = df3['review_score'] * df3['order_status']
```

```python
#Creating additional column with the mean price of products by category

df_mpc = df3.groupby('product_category_name', as_index = False).agg({'price' : np.mean}).rename(columns = {'price' : 'mean_price_cat'})

df4 = df3.merge(df_mpc, on = 'product_category_name')
```

```python
# Analysing if there are clients on the base who bought the same product more than once, giving it different reviews

df5 = df4.copy()
df5.groupby(by = ['customer_id', 'product_id']).agg({'order_id' : pd.Series.nunique})['order_id'].value_counts()

# Connclusion: no products were bought twice, on different orders, by the same costumer
```

```python
# Creating 'action' column, containing the following logic:

# 1 - Using 'order_item_id' to count how many times the product was bought. 
#     The most bought ones probably gets a more precise review, thus increasing it's 'action' column value
# 2 - Using 'review_score' to increase or decrease the 'action' value acording to the product review 
# 3 - Using 'price' to increase the value of the 'action', considering that higher value products, generally, are most critically reviewed
# 4 - Using 'mean_price_cat' to escalonate the products prices by category, to reduce the effects of category price differences


df5['action'] = df5['order_item_id'] * df5['review_score'] * df5['price'] / df5['mean_price_cat']
```

```python
# Creating a list containing the top 1000 products with the highest 'action' score

df6 = df5.copy()

df6.sort_values('action', ascending = False, inplace = True)

top_produtos = df6.head(20000)['product_id'].tolist()
```

# 4. Modeling

```python
# Creating a function to generate the cosine similarity matrix, as well as a list containing the index of the products

from sklearn.metrics.pairwise import cosine_similarity

def get_item_based_similarity_matrix(data, lista):
  
  product_sampled_relationship_data = data[data["product_id"].isin(lista)] \
    .pivot_table(index="customer_id", columns="product_id", values="action").fillna(0)

  return cosine_similarity(product_sampled_relationship_data.T), product_sampled_relationship_data.T.index

sim_matrix, available_items_index = get_item_based_similarity_matrix(df5, top_produtos)
```

```python
# Creating a function to, once a determined item is passed, return the top n values with the highest similarity as that one

def get_n_closest(item_id, n=10):

  if item_id not in available_items_index:
   raise ValueError('Item is not available on data.')
      
  item_index = available_items_index.tolist().index(item_id)
  closest_n_item = np.argsort(-sim_matrix[item_index])[1:n]
  n_closest = []

  for item_data in zip(available_items_index[closest_n_item], sim_matrix[item_index][closest_n_item]):
    n_closest.append(item_data)
  return tuple(n_closest)
```

```python
# Functioning code example

for a in list(get_n_closest('53759a2ecddad2bb87a079a1f1519f73')):
  if a[1] != 0:
    print(f'The item {a[0]} has a similarity of {a[1]*100:.2f}%')
```
